#Adathalmaz előkészítés

```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
```

```python
df
```

##Adatszétosztás X és Y halmazra

```python
y=df["logS"]
```

```python
x=df.drop('logS', axis=1)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

# Modell építése

##Linearis regresszio

```python
import sklearn.linear_model
import LinearRegression
```

```python
lr = sklearn.linear_model.LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_test_pred=lr.predict(X_test)
y_lr_train_pred=lr.predict(X_train)
```

##Kiértékelés

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mse=mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2=r2_score(y_train, y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2=r2_score(y_test, y_lr_test_pred)
```

```python

lr_results=pd.DataFrame([['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]],
                        columns=['Method','Training MSE','Training R2','Test MSE','Test R2'])
```

```python
lr_results
```

##RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(max_depth=2, random_state=42)
```

###Modell illesztés

```python
rf.fit(X_train, y_train)
```

```python
y_rf_train_pred=rf.predict(X_train)
y_rf_test_pred=rf.predict(X_test)
```

```python
rf_train_mse=mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2=r2_score(y_train, y_rf_train_pred)

rf_test_mse=mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2=r2_score(y_test, y_rf_test_pred)

rf.results=pd.DataFrame([["RandomForest", rf_train_mse, rf_train_r2,rf_test_mse, rf_test_r2 ]])
rf.results.columns=["Method", "Train MSE", "Train R2", "Test MSE", "Test R2"]
rf.results
```
