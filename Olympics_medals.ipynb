##Missing data handling

```python
teams_removed[teams_removed.isnull().any(axis=1)] #esetlegesen hiányzó adatok
```

###Adatok feltöltése, dobása

```python
teams_removed=teams_removed.dropna()
teams_removed
```

```python
train=teams_removed[teams_removed["year"]<2012].copy()
test=teams_removed[teams_removed["year"]>=2012].copy()
```

```python
train.shape ##hány sora van
```

```python
test.shape
```

#LineárisRegresszió

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train=["athletes", "prev_medals"]
y_train="medals"
```

```python
lr=LinearRegression()
lr.fit(train[X_train], train[y_train])
```

```python
predictions=lr.predict(test[X_train])
predictions
```

```python
test["predictions"]=predictions
test
```

```python
test.loc[test["predictions"]<0, "predictions"]=0
test #ahol nulla alatt van az érték nullára állítom, hiszen nem lehet kevesebb
```

```python
test["predictions"]=test["predictions"].round(0)
test
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
error=mean_absolute_error(test["medals"], test["predictions"])
error ##átlagban ennyivel lőttem mellé
```

```python
errors=(test["medals"]-test["predictions"]).abs()
errors
```

```python
error_per_team=errors.groupby(test["team"]).mean()
error_per_team
```

```python
error_per_team[error_per_team==(error_per_team.max())]
```
